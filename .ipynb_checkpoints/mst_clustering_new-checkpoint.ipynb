# Defining the euclidean distances

In the first step, we need to define euclidean distances. 

```python
def euclidean(v1, v2):
    return sum((p-q)**2 for p, q in zip(v1, v2)) ** .5
```

```python
def euclidean_distance(df):
    df = df.to_numpy()
    weight_matrix = np.zeros(shape = (len(df),len(df)))
    i = 0
    for point_x in df:
        j = 0
        for point_y in df:
            dist = np.linalg.norm(point_x - point_y)
            weight_matrix[i][j] = dist
            j += 1
        i +=1
    return weight_matrix
```

# Defining the dunn index

```python
# define dunn index since sklearn package is out of date

def δ(ck, cl):
    values = np.ones([len(ck), len(cl)])
    
    for i in range(0, len(ck)):
        for j in range(0, len(cl)):
            values[i, j] = np.linalg.norm(ck[i]-cl[j])
            
    return np.min(values)

def Δ(ci):
    values = np.zeros([len(ci), len(ci)])
    
    for i in range(0, len(ci)):
        for j in range(0, len(ci)):
            values[i, j] = np.linalg.norm(ci[i]-ci[j])
            
    return np.max(values)

def dunn(k_list):
    δs = np.ones([len(k_list), len(k_list)])
    Δs = np.zeros([len(k_list), 1])
    l_range = list(range(0, len(k_list)))
    
    for k in l_range:
        for l in (l_range[0:k]+l_range[k+1:]):
            δs[k, l] = δ(k_list[k], k_list[l])
        
        Δs[k] = Δ(k_list[k])

    di = np.min(δs)/np.max(Δs)
    return di
```

# MST_Clustering
This is the MST clustering algorithm if the dimension of the inputs are 2. For the Thyroid dataset we will need to adjust it a tiny bit and do it for 5 dimensions. 

```python
#Class 
class Node_2d:
    def __init__(self, point, p):
        # generate coordinates for higher dimension
        self.x = point[0]
        self.y = point[1]
        #self.z = point[2]
        #self.w = point[3]
        #self.q = point[4]
        self.parent = p
        self.rank = 0 # number of neighbours

#Class 

class Node_5d:
    def __init__(self, point, p):
        # generate coordinates for higher dimension
        self.x = point[0]
        self.y = point[1]
        self.z = point[2]
        self.w = point[3]
        self.q = point[4]
        self.parent = p
        self.rank = 0 # number of neighbours
        
        
class Edge:
    def __init__(self, u, v, w):
        self.u = u #endpoint of the edge
        self.v = v #the other endpoint of the edge
        self.weight = w #comes from eucledian distance
        
#to find the parent of the given node
def Find(i, nodes):
    if (i != nodes[i].parent) :
        nodes[i].parent = Find(nodes[i].parent, nodes)
    return nodes[i].parent

#to unite the two edges
def Union(u, v, nodes):
    r1 = Find(u, nodes) #parent of first point
    r2 = Find(v, nodes) #parent of second point
    if (r1 != r2):
        if (nodes[r1].rank > nodes[r2].rank):
            nodes[r2].parent = r1
        else:
            nodes[r1].parent = r2
            if (nodes[r1].rank == nodes[r2].rank):
                nodes[r2].rank += 1
                
# MST Kruskal algorithm for 2d data
def mst_kruskal_2d(df,k):
    n = df.shape[0]
    nodes = []
    for i in range(n):
        nodes.append(Node_2d(df.iloc[i], i))
        
    edges = []
    
    for i in range(n):
        point_a=df.iloc[i]
        for j in range(i+1, n):
            point_b=df.iloc[j]
            edges.append(Edge(i, j, euclidean(point_a,point_b)))
    
    edges = sorted(edges, key=lambda edge: edge.weight) #Sort the edges 
    num_edges_added = 0
    
    for edge in edges: #to iterate through the sorted edges list
        if (num_edges_added <= n-k):
            if Find(edge.u, nodes) != Find(edge.v, nodes):
                num_edges_added += 1
                Union(edge.u, edge.v, nodes)
        else:
            pass
        #if (num_edges_added > n-k):
         #   pass
    output = []
    for element in nodes:
        output.append((element.x,element.y,element.parent))
    
    output = pd.DataFrame(output, columns = [df.columns[0],df.columns[1],"label"])
    return output


# MST Kruskal algorithm for 5d data
def mst_kruskal_5d(df,k):
    n = df.shape[0]
    nodes = []
    for i in range(n):
        nodes.append(Node_5d(df.iloc[i], i))
        
    edges = []
    
    for i in range(n):
        point_a=df.iloc[i]
        for j in range(i+1, n):
            point_b=df.iloc[j]
            edges.append(Edge(i, j, euclidean(point_a,point_b)))
    
    edges = sorted(edges, key=lambda edge: edge.weight) #Sort the edges 
    num_edges_added = 0
    
    for edge in edges: #to iterate through the sorted edges list
        if (num_edges_added <= n-k):
            if Find(edge.u, nodes) != Find(edge.v, nodes):
                num_edges_added += 1
                Union(edge.u, edge.v, nodes)
        else:
            pass
        #if (num_edges_added > n-k):
         #   pass
    output = []
    for element in nodes:
        output.append((element.x,element.y,element.z, element.w, element.q, element.parent))
    
    output = pd.DataFrame(output, columns = [df.columns[0],df.columns[1],df.columns[2],df.columns[3],df.columns[4],"label"])
    return output

```

## Dunn Index 

```python
def get_dunn_2d(df):

    #mst = my_kruskal_2d(d,k)
    k_list = []
    for i in df.label.unique().tolist():
        k_list.append(df[df.label == i].values)
    return dunn(k_list)
```

# Applying the algorithms to the dataset thyroid

## Loading the dataset 

```python
thyroid = pd.read_csv('thyroid.txt', sep="  ", header=None)
thyroid.apply(pd.to_numeric)
thyroid.columns = ['a', 'b','c','d','e']
```

```python
thyroid
```

```python
from sklearn.metrics import davies_bouldin_score

print('Running Davies-Bouldin score calculations')
scores = []
k = []
for n in range(2,40):
    tdata = mst_kruskal_5d(thyroid, n)
    k.append(n)
    score = davies_bouldin_score(tdata[['a','b','c','d','e']], tdata.label)
    scores.append(score)

sns.lineplot(x=k,y=scores).set_title('Davies-Bouldin Score of K Clusters')

print(f'Best score is with {k[scores.index(min(scores))]} clusters :{min(scores)}')
```

# Importing the quality measures using dunn index

```python
def get_dunn(d,k):
    output = mst_kruskal_5d(d,k)
    k_list = []
    for i in output.label.unique().tolist():
        k_list.append(output[output.label == i].values)
    return dunn(k_list)
```

```python
scores = []
k = []
print('Running Dunn Score calculations: ')
for n in range(2,50):
    score = get_dunn(thyroid,n)
    k.append(n)
    scores.append(score)

sns.lineplot(x=k,y=scores).set_title('Dunn Score of K Clusters')
print(f'Best score is with {k[scores.index(max(scores))]} clusters :{max(scores)}')
```
